```python
edades = list(censo['Edad'])[1:]
tramos_edad = []
for tramo in edades:
    if len(tramo) == 5:
        num = int(tramo[-1])
    elif "100" in tramo:
        num = 100
    elif "Total Comunal" in tramo:
        num = -1
    else:
        num = int(tramo[-1] + tramo[-2])
    if 0 <= num <= 19:
        tramos_edad.append("0 a 19 años")
    elif num == -1:
        tramos_edad.append("-")
    elif num > 19:
        tramos_edad.append("20 o más años")
```

```python
tramos_edad = ["-"] + tramos_edad
```

```python
censo['Tramo Edad'] = tramos_edad
```
